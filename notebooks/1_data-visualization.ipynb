```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 8
```

```python
mpg_df = pd.read_csv("../data/auto_mpg_preprocessed.csv")
occup_df = pd.read_csv("../data/occupancy_detection_preprocessed.csv")
```

```python
mpg_df = mpg_df.drop(columns="Unnamed: 0")
occup_df = occup_df.drop(columns="Unnamed: 0")
```

```python
sns.pairplot(occup_df, hue='Occupancy')
```

```python
sns.boxplot(x='Occupancy', y='Temperature', data=occup_df)
```

```python
sns.boxplot(x='Occupancy', y='Humidity', data=occup_df)#%%
```

```python
sns.boxplot(x='Occupancy', y='HumidityRatio', data=occup_df)
```

```python
sns.boxplot(x='Occupancy', y='CO2', data=occup_df)
```

```python
sns.boxplot(x='Occupancy', y='Light', data=occup_df)
```

По данным графикам видно, что значения атрибутов в основном выше, когда в комнате присутствует человек.

```python
occup_df.plot(x='date', y=['Light','CO2', 'Humidity', 'Temperature', 'Occupancy'], subplots=True, kind="area")
```

По данному графику видно, что значения атрибутов повышаются при присутствии человека в комнате<br>
Особенно это заметно на признаках "Light" и "CO2".

___

```python
mpg_df.plot(y=mpg_df.drop(columns=['model year', 'origin', 'acceleration']).columns, subplots=True)
```

Из данного графика видно, что значение ключевого атрибута повышается, когда значение представленных атрибутов ниже.

```python
mpg_df.plot(y=['mpg', 'acceleration'], subplots=True)
```

Из данного графика видно, что значения ключевого атрибута возрастает вместе со значением атрибута "acceleration".

```python
sns.boxplot(x='origin', y='mpg', data=mpg_df)
```

```python
sns.boxplot(x='model year', y='mpg', data=mpg_df)
```

Данные графики показывают распределение ключевого атрибута на атрибутах "origin" и "model year"
Видно, что машины из Европы(2) и Азии(3) имеют 'mpg' выше, чем машины из Америки(1)

С годом производства не всё так просто, но можно сказать что модель более позднего года, вероятно, будет иметь более высокое значение 'mpg', чем модели старше.
