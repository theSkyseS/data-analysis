```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 8
```

```python
mpg_df = pd.read_csv("../data/auto_mpg_preprocessed.csv")
wine = pd.read_csv("../data/wine_preprocessed.csv")
```

```python
mpg_df = mpg_df.drop(columns="Unnamed: 0")
wine = wine.drop(columns="Unnamed: 0")
```

```python
sns.boxplot(x='Cultivar', y='Alcohol', data=wine)
```

Из данного графика видно, что сорт 2 содержит, в основном, меньше алкоголя, чем 3 и 1

```python
sns.boxplot(x='Cultivar', y='Malic acid', data=wine)
```

Из данного графика видно, что вино третьего сорта, содержит больше яблочной кислоты, чем 1 и 2.
При этом содержание яблочной кислоты в винах первого сорта лежит в узком интервале.

```python
sns.boxplot(x='Cultivar', y='Ash', data=wine)#%%
```

По этому графику можно сказать, что вина всех сортов содержат примерно одинаковое количество золы.

```python
sns.boxplot(x='Cultivar', y='Alcalinity of ash', data=wine)
```

Данный график показывает, что щёлочность золы растёт вместе с сортом (т.е. сорт 2 имеет значения выше, чем 1, а сорт 3 - чем 2)

```python
sns.boxplot(x='Cultivar', y='Magnesium', data=wine)
```

Из данного графика можно сделать вывод, что содержание магния в винах 2го и 3го сорта ниже, чем в винах 1го.

```python
mpg_df.plot(y=mpg_df.drop(columns=['model year', 'origin', 'acceleration']).columns, subplots=True)
```

Из данного графика видно, что значение ключевого атрибута повышается, когда значение представленных атрибутов ниже.

```python
mpg_df.plot(y=['mpg', 'acceleration'], subplots=True)
```

Из данного графика видно, что значения ключевого атрибута возрастает вместе со значением атрибута "acceleration".

```python
sns.boxplot(x='origin', y='mpg', data=mpg_df)
```

```python
sns.boxplot(x='model year', y='mpg', data=mpg_df)
```

Данные графики показывают распределение ключевого атрибута на атрибутах "origin" и "model year"
Видно, что машины из Европы(2) и Азии(3) имеют 'mpg' выше, чем машины из Америки(1)

С годом производства не всё так просто, но можно сказать что модель более позднего года, вероятно, будет иметь более высокое значение 'mpg', чем модели старше.
